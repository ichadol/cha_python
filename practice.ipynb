```python
df_cospi = pd.read_csv('https://raw.githubusercontent.com/jin0choi1216/dataset/main/KOSPI_stocks.csv', index_col= 0)
df_cospi.head()
```

```python
df_cospi['high_level'] = df_cospi['High'].rank(ascending=False)
df_cospi.sort_values('high_level', ascending= True).reset_index()
```

```python
def high_check(a):
    if a >= 400000:
        return "높은 주"
    elif a < 400000 and a >= 200000:
        return "중간 주"
    else:
        return "낮은 주"
    
df_cospi['주가 비교'] = df_cospi['High'].apply(high_check)
df_cospi
```

```python
df_cospi['주가 비교 두 번째'] = df_cospi['High'].apply(lambda x : '높은 주' if x > 400000 
                                                else ('중간 주' if x >= 200000 and x <= 400000 else '낮은 주'))
df_cospi
```

```python
df_cospi['주가 비교'].value_counts()
```
