```python
import pandas as pd
df_krx = pd.read_csv("https://raw.githubusercontent.com/jin0choi1216/dataset/main/KRX_stocks.csv", index_col=0)
df_krx.head()
```

```python
df_krx['Marcap_rank'] = df_krx['Marcap'].rank(ascending= False)
df_krx
```

```python
def small(a):
    if a <= 100:
        return "large-cap"
    elif a > 100 and a <= 300:
        return "mid-cap"
    else:
        return "small-cap"

df_krx['Marcap_size'] = df_krx['Marcap_rank'].apply(small)
df_krx
```

```python
df_krx[['Marcap_size', 'ChagesRatio']].groupby('Marcap_size').mean()
```
