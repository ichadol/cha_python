```python

dates = pd.date_range('2023-09-11', periods= 3)
prices = {'삼전': [40000, 50000, 60000], '엘전': [100000, 200000, 3000000], '하닉': [1000, 2000, 3000]}
df_jusik = pd.DataFrame(prices, index = dates)
df_jusik
```

```python
s1 = pd.Series([211500, 215000, 216500], index = dates, name = " 네이버")
concat_dfs1 = pd.concat([df_jusik, s1], axis = 1)
concat_dfs1['카카오'] = concat_dfs1['하닉'] + 1000
concat_dfs1.loc['2023-09-10',:] = concat_dfs1.loc['2023-09-11',:] + 20000
concat_dfs1.sort_index()
concat_dfs1.sort_values(by = '엘전', ascending=False)
```

```python
concat_dfs1.drop('삼전', axis = 1)
concat_dfs1.sort_index()
concat_dfs1.sort_values("하닉")

concat_dfs1.loc['2023-09-20',:] = concat_dfs1.loc['2023-09-11',:] + 21300
concat_dfs1

concat_dfs1['모든주 더하기'] = concat_dfs1['삼전'] + concat_dfs1['엘전'] + concat_dfs1['하닉'] + concat_dfs1['카카오'] 

```

```python
concat_dfs1.sort_index()

a = concat_dfs1.drop("모든주 더하기", axis = 1)

a.sort_index()

a['sum'] = a.sum(axis = 1)
a

a['mean'] = a.mean(axis = 1)
a

a['man'] = a.mean(axis = 0)

a.loc['2023-09-11': '2023-09-20', 'man'] = 10


a.fillna(method = "ffill")

a.interpolate()
```
